```python

#Python Simulation Of SINR (Signal to Interference + Noise Ratio )
import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0,100,1)
signal_power = np.random.normal(0, 2, len(time)) + 15 #15 dB
interference = np.random.normal(0, 1, len(time)) + 5 #5 dB
noise = np.random.normal(0, 0.5 , len(time)) + 2 # 2 dB


sinr = signal_power/(interference + noise)
plt.plot(time, 10*np.log10(sinr))
plt.xlabel('Time (s)')
plt.ylabel('SINR (dB)')
plt.title('Simulated SINR over Time')
plt.grid(True)
plt.show()
```
